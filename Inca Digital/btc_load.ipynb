```python
import pandas as pd
import datetime as dt
import numpy as np
import json
from pprint import pprint
```

```python
###загрузка и предобработка данных о транзакциях биткоина
```

```python
%%time
file1 = open("download_data_fincen_files/btc-tx.csv", "r")
line = file1.readline()

dict_ = {}
i = 0
err = 0
err1 = 0

while True:
    line = file1.readline()
    if not line:
        break

    line_clear = line.replace('"', '').replace('\n', '')
    
    if len(line_clear) > 0:
        ll = line_clear.split(',')
    
        try:
            dict_[i] = {'time': ll[0], 
                        'hash': ll[1], 
                        'sender': ll[2], 
                        'receiver': ll[3], 
                        'BTC': ll[4]}
            i = i + 1
        
        except IndexError:
            err1 = err1 + 1
            if dict_[i-1]['BTC'] == '':
                dict_[i-1]['BTC'] = ll[0].replace(' ', '')
            elif dict_[i-1]['BTC'].replace('.', '').isdigit():
                lines = []
                comma_count = 0
                while comma_count == 0:
                    lines.append(file1.readline().replace('"', '').replace('\n', ''))
                    comma_count = lines[-1].count(',')
                last_str = lines[-1].split(',')
        
                dict_[i] = {'time': ll[0], 
                            'hash': ll[1], 
                            'sender': ll[2] + ' ' + last_str[0] + ' ' + ' '.join([item for item in lines[:-1]]), 
                            'receiver': last_str[1], 
                            'BTC': last_str[2]}
                i = i + 1
   
            else:
                print(f'строка {i}')
                break
            
            continue
    
    else:
        err = err + 1

# закрываем файл
file1.close
```

```python
btc_exch_rate = pd.read_csv('download_data_fincen_files/Bitcoin_Historical_Data_Investing.com.csv')
```

```python
def price_convert(price):
    return price.replace(',', '')
btc_exch_rate['Price'] = btc_exch_rate['Price'].apply(price_convert) #надо убрать запятую, иначе не конвертится

btc_exch_rate['Date'] = btc_exch_rate['Date'].astype('datetime64')
btc_exch_rate['Price'] = btc_exch_rate['Price'].replace(',', '').astype('float64')
```

```python
#берем банковскую транзакцию и ищем соответсвующую принятым допущениям транзакцию биткоинов
#проблема - дата банковской транзакции описана периодом дат (от одного дня)
```

```python
%%time
bank_trans_dict = {}
i = 0

#перебираем все банковские транзакции
for trans in trans_map.index:
    #счетчик банковских транзакций
    i = i + 1 
    
    #получаем дату для текущей банковской транзакции
    trans_date_begin = trans_map.iloc[trans]['begin_date'] #период банковской транзакции (может быть неск.транзакций)
    trans_date_end = trans_map.iloc[trans]['end_date']
    
    #средняя сумма банковских транзакций в этот период
    trans_amount = trans_map.iloc[trans]['amount_transactions'] / trans_map.iloc[trans]['number_transactions']
    
    #подсчет курса BTC: берем средний курс 1)за 4 дня при периоде в 1 день, 
    #2) за период 3 дня до начальной даты по конечную дату
    btc_rate = round(btc_exch_rate[(btc_exch_rate['Date'] >= trans_date_begin - dt.timedelta(days = 3)) &\
                                   (btc_exch_rate['Date'] <= trans_date_end)]['Price'].mean(), 5)
   
    trans_amount_btc = trans_amount / btc_rate #сумма транзакций в эту дату в биткоинах
        
    #теперь ищем "похожие" транзакции биткоинов
    date_range = trans_date_begin - dt.timedelta(days = 3)
    amount_range = trans_amount_btc * 1.01
        
    #выборка записей по транзакциям биткоина по сформированным условиям
    df_btc_trans = btc[((btc['time'] >= date_range) & (btc['time'] <= trans_date_end)) & \
                        ((btc['BTC'] >= trans_amount_btc) & (btc['BTC'] <= amount_range))]
    
    #заполняем словарь
    if len(df_btc_trans) > 0:
        for ind, row in df_btc_trans.iterrows():
            #подготавливаем служебное наполнение
            delta_date = trans_date_end - row['time']
            
            delta_amount_perc = \
                round((row['BTC'] - trans_amount_btc) * 100 / trans_amount_btc, 4)
            
            if bank_trans_dict.get(trans, None) == None:
                bank_trans_dict[trans] = \
                {
                    'date_begin' : dt.datetime.strftime(trans_date_begin, '%Y-%m-%d'),
                    'date_end' : dt.datetime.strftime(trans_date_end, '%Y-%m-%d'),
                    'amount_in_dollars' : trans_amount,
                    'amount_in_btc' : trans_amount_btc,
                     ind : {
                                'date' : dt.datetime.strftime(row['time'], '%Y-%m-%d'),
                                'amount' : row['BTC'],
                                'delta_date' : delta_date.days, 
                                'delta_amount_perc' : delta_amount_perc
                        }
                }
            else:
                bank_trans_dict[trans][ind] = \
                {
                    'date' : dt.datetime.strftime(row['time'], '%Y-%m-%d'),
                    'amount' : row['BTC'],
                    'delta_date' : delta_date.days, 
                    'delta_amount_perc' : delta_amount_perc
                }
```
