```python
import pandas as pd
import datetime as dt
import numpy as np
import json
from pprint import pprint
```

```python
with open('download_data_fincen_files/data.txt', 'r') as fp:
    data = json.load(fp)
```

```python
i=0
j=0
for bank_trans in data.items():
    i=i+1
    if bank_trans[1]['date_begin'] == bank_trans[1]['date_end']:
        j=j+1
        print(f'Банковская транзакция: {bank_trans[0]}')
        print(f"\tпериод: {bank_trans[1]['date_begin']} / {bank_trans[1]['date_end']} \
            \n\tср.сумма $: {round(bank_trans[1]['amount_in_dollars'],2)} \
            \n\tср.сумма BTC: {round(bank_trans[1]['amount_in_btc'],2)}")
 
        for item in bank_trans[1]:
            try:
                int(item)
                btc_tr = bank_trans[1][item]
                print(f'\t\tтранзакция BTC: {item}')
                print(f"\t\tдата: {btc_tr['date']}, сумма BTC: {round(btc_tr['amount'],2)}, \
                    \n\t\tразница дат: {btc_tr['delta_date']}, разница сумм %: {btc_tr['delta_amount_perc']}")
            except ValueError:
                continue
    
        print(f'конец банковской транзакции: {bank_trans[0]}\n')
```
